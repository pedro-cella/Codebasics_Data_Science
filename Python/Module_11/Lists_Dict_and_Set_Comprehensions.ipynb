# Lists, Dict and Set Comprehensions
- Not only List Comprehension are shorter to write but the have a better performance as well

```python
numbers = [1, 2, 3, 4, 5]
sqr_numbers = []

for n in numbers:
    sqr_numbers.append(n*n)

print(sqr_numbers)
```

```python
sqr_numbers = [n*n for n in numbers]
sqr_numbers
```

```python
import time

long_list = range(10000000)
sqr_list = []

start= time.time()
for n in long_list:
    sqr_list.append(n*n)
end = time.time()
print("Time taken by for loop:", (end-start))
```

```python
start = time.time()
sqr_list = [n*n for n in long_list]
end = time.time()
print("Time taken by list comprehension:", (end-start))
```

```python
price = [100, 120, 80, 50]
earning_per_share = [9, 15, 40, 2]
```

```python
tuple(zip(price, earning_per_share))
```

```python
pe = []

for p, e in zip(price, earning_per_share):
    pe.append(p/e)
pe
```

```python
pe = [p/e for p, e in zip(price, earning_per_share)]
pe
```

```python
price = [100, 120, 80, 50]
earning_per_share = [0, 15, 40, 2]

pe = [p/e if e != 0 else -1 for p, e in zip(price, earning_per_share)]
pe
```

```python
words = ["apple", "banana", "cherry", "date"]

lengths = [len(word) for word in words]
lengths
```

```python
prices = [120, 99, 150, 80, 200]

new_prices = [price*0.9 if price > 100 else price for price in prices]
new_prices
```

```python
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]

name_age_dict = {names[i]: ages[i]for i in range(len(names))}
name_age_dict
```

```python
transactions = [
    {'customer_name': 'Mira', "amount": 56, "date": "2024-08-01"},
    {'customer_name': 'Mira', "amount": 12.3, "date": "2024-09-05"},
    {'customer_name': 'Mohan', "amount": 21, "date": "2024-09-03"},
    {'customer_name': 'Mohan', "amount": 36.8, "date": "2024-09-10"},
    {'customer_name': 'Lucy', "amount": 66.3, "date": "2024-09-10"},
]
```

```python
customers = set()

for trans in transactions:
    customers.add(trans['customer_name'])

customers
```

```python
customers = {trans['customer_name'] for trans in transactions}
customers
```

## Takeaways
- **List comprehension** provide a concise way to create lists by running a for loop in a single-line.
- Comprehensions are generally **more readable** and **faster than using loops** for building collections, making them a preferred choic for such tasks in Python
- **Dictionary comprehensions** allow for the dynamic construction of dictionaries by running a for loop in a single-line
- **Set comprehension** work similarly to list comprehensions bu produced a set, automatically removing dupllicate elements during creation.
- **All comprehensions** can include conditionals to filter **elements, providing a powerful tool** for creating customized collections efficiently.
