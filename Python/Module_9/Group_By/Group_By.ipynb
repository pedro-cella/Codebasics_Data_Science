# Group By

```python
import pandas as pd
```

```python
df = pd.read_csv('weather_by_cities.csv')
df
```

```python
df[df.city=="new york"].temperature.max()
```

```python
g = df.groupby("city")
g
```

```python
for city, data in g:
    print("city:", city)
    print("\n")
    print("data:", data.temperature.max())
```

```python
def grouper(df, idx, col):
    if 90 <= df[col].loc[idx] <= 90:
        return '80-90'
    elif 50 <= df[col].loc[idx] <= 60:
        return '50-60'
    else:
        return 'others'
```

```python
g = df.groupby(lambda idx: grouper(df, idx, 'temperature'))
g
```

```python
for key, d in g:
    print("Key:", key)
    print("Data:", d)
```
