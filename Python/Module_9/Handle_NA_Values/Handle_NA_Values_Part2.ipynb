# Handle NA Values Part 2

```python
import pandas as pd
```

```python
df = pd.read_csv('weather_data-Copy1.csv')
df
```

```python
import numpy as np
df.replace([-99999, -88888], value=np.nan)
```

```python
df
```

```python
df.replace({
    'temperature': -99999,
    'windspeed': [-99999, -88888],
    'event': r'^\s*no event\s*$'
}, np.nan, regex=True)
```

```python
df.replace({
    -99999: np.nan,
    -88888: np.nan,
    r'^\s*no event\s*$': 'Sunny'
}, regex=True)
```

```python
df = pd.DataFrame({
    'score': ['exceptional', 'average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df
```

```python
df.replace(['poor', 'average', 'good', 'exceptional'], [1,2,3,4])
```
