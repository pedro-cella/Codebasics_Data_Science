# Handle NA Values Part 1

```python
import pandas as pd

df = pd.read_csv('weather_data.csv', parse_dates=["day"])
df
```

```python
type(df.day[0])
```

```python
df.set_index('day', inplace=True)
df
```

```python
df.fillna({'temperature': 0, 'windspeed': 0, 'event': '0'}, inplace=True)
df
```

```python
df
```

```python
df.temperature.mean()
```

```python
df.fillna({
    'temperature': df.temperature.mean(),
    'windspeed': df.windspeed.mean(),
    'event': 'No Event'
})
```

```python
df
```

```python
df.ffill()
```

```python
df
```

```python
df.bfill()
```

```python
df
```

```python
df.bfill(axis='columns')
```

```python
df.ffill(limit=2)
```

```python
df
```

```python
df.interpolate()
```

```python
df.dropna(how='all')
```

```python
df.dropna(thresh=2)
```

## Takeaways
- Handling the **missing/Null values** in the dataset is a core responsability of **Data Analyst/Data Scientist.**
- Deciding how to handle the **Na/NaN values** tottally depends on the business context. You can either use the **mean, median, ** or **previous day value**, depending on the discussion with the stakeholders.
