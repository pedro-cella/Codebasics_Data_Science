# `__main__` Functions

## Takeaways
- **`if __name__ == "__main__`"** is a way to define entry point in Python code (similar to main function in C++ or JAVA)
- When you import any module (e.g. import NumPy as np), the module.`__name__` is set to the name of that module (e.g. np.`__name__` is set as 'Numpy')
