# Exception Handling

## Takeaways
- **Exception handling** ensures that programs can adress and recover from errors during execution without crashing
- Using **try and except blocks** allows developers to separate normal code from error handling code, **enhancing readability**
- Specifying particular excpetions to catch enables **targeted responses to different error types,** improving error resolution effectiveness.
- The **finally block** executes code regardless of wether an exception was raised or not, ensuring that cleanup and release of resources can always occur
