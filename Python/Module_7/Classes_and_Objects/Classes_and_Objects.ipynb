# Classes and Objects

## Takeaways
- **Object-oriented programming (OOPS)** makes managing software projects easier.
- **Classes** and **objects** are at the core of **OOPS**
- Classes are like a **blueprint of an entity** whereas objects represent a **specific instance of that class.** For example Human is a class whereas specific human being (say Mira Sharma) is an instance of the class Human.
- Classes have **properties** and **functions.**
- **Use pip install package_name** to add new packages to your Python environment.
- Use the **__init__** method to initialize object properties when an object is created.
- Methods defined in a class describe the behaviors that objects of that class can perform.
