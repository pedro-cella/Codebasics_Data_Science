# Matrix Operations

```python
import numpy as np
```

```python
# Sales data for Quarter 1 (Matrix 1)
# Rows represent different products, columns represent different regions

q1 = np.array([
    [200, 220, 250], # Product A
    [150, 180, 210], # Product B
    [300, 330, 360], # Product C
])

q2 = np.array([
    [209, 231, 259], # Product A
    [155, 192, 222], # Product B
    [310, 340, 375]  # Product C
])
```

```python
q1 + q2
```

```python
q2 - q1
```

```python
(q2-q1)*100/q1
```

prices = np.array([
    [10, 12, 11], # Prices for Product A in diffeent regions
    [8, 9, 10],   # Prices for Product B
    [15, 16, 17],
])

q1_revenue = q1 * prices
q1_revenue

```python
q1_discounts = q1_revenue*0.2
q1_discounts
```

```python
net_revenue = q1_revenue - q1_discounts
net_revenue
```

```python
np.sum(q1_discounts)
```

```python
np.sum(net_revenue)
```

```python
features = np.array([
    [2000, 3], # house 1
    [1800, 2]  # house 2
])

weights = np.array([150, 50000])

np.dot(features, weights)
```

```python
c = np.array([1, 2, 3])
d = np.array([4, 5, 6])
np.cross(c, d)
```

## Takeaways
- **Using Numpy,** one can perform matrix operations in a convenient way.
- **For two matrices (m1 and m2),** one can do **addition, subtraction, division** and **multiplication** by using **+.-./** and ** * operators  **
- **np.dot(m1, m2)** and **np.cross(m1, m2)** can be used to perform a dot and cross products between two matrices (m1 and m2)
