```python
class RemoteControl:
    def __init__(self):
        self.channels = ["NDTV", "CNN", "HBO"]
        self.index = -1

    def __next__(self):
        self.index += 1
        if self.index == len(self.channels):
            raise StopIteration
            
        return self.channels[self.index]

    def __iter__(self):
        return self
```

```python
rc = RemoteControl()
for channel in rc:
    print(channel)
```

```python
rc = RemoteControl()

next(rc)
```

```python
def remote_control():
    yield "NDTV"
    yield "CNN"
    yield "ESPN"
```

```python
rc = remote_control()
type(rc)
```

```python
for channel in rc:
    print(channel)
```

```python
def get_company_data():
    for i in range(3):
        with open(f"stock_{i+1}.txt", "r") as f:
            content = f.read()
            yield content
```

```python
for company_data in get_company_data():
    print(company_data)
```

```python
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a+b
```

```python
fib_gen = fibonacci()

for _ in range(10):
    print(next(fib_gen))
```

## Takeaways
- **Generators in Python are a simples way to create iterators using functions** that yield values instead of returning a single value.
- Iterators are objects that implement the **__iter__() and __next__() methods,** which allow Python to iterate over collections of items, such as in a loop.
- Generators help in **managing memory efficiently by yielding items one at a time,** only holding one item in memory, unlike lists which store all elements.
- The **yield statement** is used in a function to turn it into a generator, suspending the function's state until the next value is needed.
- Generators and iterators are powerful for handling large data sets, infinite squences, and pipelines that transform data through a series of steps.
