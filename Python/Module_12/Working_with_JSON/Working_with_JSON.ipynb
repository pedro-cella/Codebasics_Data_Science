# Working with JSON

```python
# JSON string
json_string = """
{
    "name": "Maulik",
    "age": 30,
    "is_employee": true,
    "skills": ["Pythn", "Data Analysis", "Machine Learning"],
    "address": {
        "street": "123 Main St",
        "city": "Wonderland"
    }
}
"""
type(json_string)
```

```python
import json

record = json.loads(json_string)
record
```

```python
type(record)
```

```python
record['name']
```

```python
data = {
    "name": "Bob",
    "age": 25,
    "is_employee": False,
    "skills": ["JavaScript", "Web Development"],
    "address": {
        "street": "456 Elm St",
        "city": "Springfield"
    }
}
```

```python
data_str = json.dumps(data)
data_str
```

```python
type(data_str)
```

```python
with open("output.json", "w") as f:
    json.dump(data, f, indent=4)
```

```python
with open('products.json', 'r') as file:
    products = json.load(file)
products
```

```python
type(products)
```

```python
products[0]['price']
```

```python
json_string = '''
{
    "books": [
        {"title": "The Magic Of Thinking Big", "author": "David J. Schwartz", "published": 1959},
        {"title": "Chanakya Neeti", "author": "B.K. Chaturvedi", "published": 2009},
        {"title": "How to Win Friends and Influence People", "author": "Dale Karnegie", "published": 1936}
    ],
    "status_code": 200
}
'''
```

```python
data = json.loads(json_string)
data
```

```python
books = data['books']
books
```

```python
books_list = [{'title': book['title'], 'author': book['author']} for book in books]
books_list
```

```python
# Malformed JSON string
json_string = """
{
    "name": "Krishna",
    "age": 30
    "is_employee": true
}
"""
```

```python
json.loads(json_string)
```

```python
try:
    json.loads(json_string)
except json.JSONDecodeError as e:
    print(f"Failed to parse JSON: {e}")
    
```

## Takeaways
- **JSON(JavaScript Objet Notation)** is a lightweight data format widely used for data exchange between servers and web applications.
- Python provides the **json** module to encode Python objects into JSON format using **json.dumps()** and to decode JSON into Python objects using **json.loads()**
- When working with files, use **json.dump()** to serialize an object to a file and **json.load()** to deserialize a file containing JSON data into a Python object.
- JSON in Python supports basic data types like **strings, numbers, lists,** and **dictionaries, making it a versatile for data storage** and **communication needs.**
- Handling exceptions like **JSONDecodeError** is essential for robust JSON data parsing, especially when dealing with data from untrusted sources.
