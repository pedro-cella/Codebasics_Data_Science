## Takeaways
- **Lists** alllow you to store sequential data.
- **Lists are ordered,**, meaning each item has a fixed position **unless explicitly changed.**
- Python lists can hold different data types in a single list, including numbers, strings, and other lists. This means they are heterogeneous. For example **my_list = ["car", 4.5, True]**
- **Lists are mutable,** allowing for elements to be **added, removed,** or **changed** within the same list.
- List slicing lets you access a specific range of elements quickly, using the syntax **list[start:end:step]**
