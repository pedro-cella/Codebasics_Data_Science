# For Loop
- You can use **else** after an **for loop**;
- You can use **zip()** to combine two or more sequences, you can use either **range()** or **enumerate()** as well

````python
# expenses

expenses = [1200, 1300, 1500, 1700]
total_expense = expenses[0] + expenses[1] + expenses[2] + expenses[3]
print(total_expense)
````

````python
total_expense = 0
for expense in expenses:
    total_expense = total_expense + expense

print(total_expense)
````

````python
total_expense = 0

# for i in range(len(expenses)):
#     expense = expenses[i]
#     print(f"Month {i+1}, Expense: {expense}")
#     total_expense += expense

# for encurtado
for i, expense in enumerate(expenses):
    print(f"Month {i+1}, Expense: {expense}")
    total_expense += expense

print(f"Total: {total_expense}")
````

````python
# break_in_for_loop

monthly_sales = [42, 38, 33, 38, 40, 45]
months = ["Jan", "Feb", "March", "April", "May", "June"]

threshold = 35

for sales_amount, month in zip(monthly_sales, months):
    if sales_amount < threshold:
        print(f"Sales amount {sales_amount} is less than the threshold in {month}")
        break
    else:
        print(f"Sales amount {sales_amount} is greater than the threshold in {month}")
````

````python
# continue_in_for_loop

for i in range(1, 11):
    if i % 2 == 0:
        continue
        
    print(i)
````

````python
n = 0

while n < 10:
    print(n)
    n += 1 
````

````python
# nested_for_loop
products = ["IPhone", "IPad", "Macbook"]
regions = ["USA", "China", "India"]
revenue = [20, 23, 45, 18, 17, 12, 12, 9, 5]

i = 0
for product in products:
    for region in regions:
        rev = revenue[i]
        i = i + 1
        print(f"{product} {region} revenue: {rev}")
````

````python
for i in range(5):
    print(i)
else:
    print("For loop terminated")
````

## Takeaways
- **for loops iterate over a sequence,** such as a list, tuple, or string, executing a block of code for each item.
- The **range() function** is often used with for loops to generate a sequence of numbers, facilitating iteration ver a set number os steps.
- Loop control statements like **break** and **continue** can alter the flow of a loop, with break eiting the loop and continue skipping to the next iteration.
- **enumerate()** will allow you to access both the inde and the element from a list inside the for loop.

Using `zip()`

```python
for sales_amount, month in zip(monthly_sales, months):
    if sales_amount < threshold:
        print(f"Sales amount {sales_amount} is less than the threshold in {month}")
        break
    else:
        print(f"Sales amount {sales_amount} is greater than the threshold in {month}")
```

Using `range()`
```python
for i in range(len(monthly_sales)):
    sales_amount = monthly_sales[i]
    month = months[i]

    if sales_amount < threshold:
        print(f"Sales amount {sales_amount} is less than the threshold in {month}")
        break
    else:
        print(f"Sales amount {sales_amount} is greater than the threshold in {month}")
```

Using `enumerate()`
```python
for i, sales_amount in enumerate(monthly_sales):
    month = months[i]

    if sales_amount < threshold:
        print(f"Sales amount {sales_amount} is less than the threshold in {month}")
        break
    else:
        print(f"Sales amount {sales_amount} is greater than the threshold in {month}")
```
