# What is an API ?
- 4 types of request:
    - GET: Read data
    - POST: Crete data
    - PUT: Update data
    - Delete: Delete data

## Takeaways
- **API** stands for **Application Programming Interface.**
- It is a way through which different software programs talk to each other.
- requests is a popular Python package used to **call APIs** from **Python code.**
- **FastAPI** is a popular web framework in Python that is used to build APIs.
- **GET, PUT, POST** and **DELETE** are four frequently used **REST API calls.**
