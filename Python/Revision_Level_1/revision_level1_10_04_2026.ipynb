## ⚔️ Missão 6 — Simulação de uso do pip

Responde (sem código):

- O que o pip faz?
  **O pip é um gerencidador de pacotes do Python, ele serve para buscar, instalar e desinstalar módulos Python**
- Qual comando instala uma biblioteca?
  **pip install `biblioteca`**
- Qual comando mostra todas instaladas?
  **- pip list → lista de forma mais legível**
  **- pip freeze → mais usado para gerar requirements.txt**
- Qual comando remove uma biblioteca?
  **pip uninstall `biblioteca`**
