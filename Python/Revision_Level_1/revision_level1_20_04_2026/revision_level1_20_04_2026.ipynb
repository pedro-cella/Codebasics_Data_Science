- Revision about the starting fundamentals of handling Null/NaN values

```python
import pandas as pd
```

## Nível 1 — Fundamentos (1–4)

## ⚔️ 1.

Mostre:

👉 quantos valores nulos existem em cada coluna

```python
df = pd.read_csv("weather_data.csv")
df
```

```python
df.isna().sum()
```

## ⚔️ 2.

Mostre:

👉 apenas as linhas que possuem pelo menos 1 valor nulo

```python
df[df.isnull().any(axis=1)]
```

## ⚔️ 3.

Mostre:

👉 apenas as linhas onde temperature está nulo

```python
df[df.temperature.isnull()]
```

## ⚔️ 4.

Descubra:

👉 quantos valores nulos existem no dataset inteiro

```python
df.isnull().sum().sum()
```

## Nível 2 — Preenchimento (5–7)

## ⚔️ 5.

Preencha os valores nulos de:

- temperature com 0
- windspeed com 0
- event com "No Event"

👉 (sem alterar o df original)

```python
df_copy = df.copy()
```

```python
df_copy.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': "No Event"
}, inplace=True)
df_copy
```

## ⚔️ 6.

Preencha os valores nulos de:

👉 temperature com a média da coluna

```python
df_copy.fillna({
    'temperature': df.temperature.mean()
}, inplace=True)
df_copy
```

## ⚔️ 7.

Preencha os valores nulos usando:

👉 o valor anterior (forward fill)

```python
df_copy.ffill()
```

## Nível 3 — Manipulação mais esperta (8–10)

## ⚔️ 8.

Preencha os valores nulos usando:

👉 o valor seguinte (backfill)

```python
df_copy.bfill()
```

## ⚔️ 9.

Remova:

👉 linhas onde TODOS os valores são nulos

```python
df_copy = pd.read_csv("weather_data.csv", parse_dates=['day'])
df_copy.set_index('day', inplace=True)
df_copy.dropna(how='all', inplace=True)
df_copy
```

## ⚔️ 10.

Remova:

👉 linhas que tenham menos de 2 valores preenchidos

```python
df_copy = df.copy()
df_copy.dropna(thresh=2, inplace=True)
df_copy
```
