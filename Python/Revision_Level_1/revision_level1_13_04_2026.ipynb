## ⚙️ Parte 2 — Atributos do Array

```python
# Crie um array e mostre:
#- shape
#- tamanho (size)
#- tipo de dado (dtype)

array = np.arange(20, 30)
print(array)
print(array.shape)
print(array.size)
print(array.dtype)
```

# 🏹 Parte 3 — Indexação e Slicing

```python
# Acesse o primeiro elemento de um array.

array = np.arange(0, 30)
array[0]
```

```python
# Acesse o último elemento de um array.
array[-1]
```

```python
# Pegue os 3 primeiros elementos de um array.
array[:3]
```

```python
# Pegue os elementos do meio (ex: índice 2 até 5).
array[14:19]
```

```python
# Em uma matriz 3x3:
# - pegue o elemento da linha 2, coluna 3

matrix = np.arange(1, 10).reshape(3, 3)
print(matrix)
print(matrix[1, 2])
```

## 🧪 Parte 4 — Operações Básicas

```python
# Crie dois arrays e faça:
# - soma
# - subtração

array1 = np.arange(0, 10)
array2 = np.arange(11, 21)
print(array1)
print(array2)

# Sum
array_soma = array1 + array2
print(array_soma)

# Sum
array_subtrac = array2 - array1
print(array_subtrac)
```

```python
# Multiplique todos os elementos de um array por 2.
array_multiply_by_2 = np.arange(0, 10)*2
array_multiply_by_2
```

```python
# Faça a média de um array.
print(array_multiplly_by_2.mean())
```

```python
# Encontre:
# - valor máximo
# - valor mínimo

print(array.max())
print(array.min())
```

```python
# Crie duas matrizes 2x2 e faça:
# - soma entre elas

matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[78, 34], [23, 14]])
print(matrix1 + matrix2)
```

```python
# Faça multiplicação de matriz (dot product).
print(np.dot(matrix1, matrix2))
```

```python
# Transponha uma matriz.
print(matrix1)
print(matrix1.T)
```

```python
# Crie um array com números aleatórios de 0 a 1.
random_array = np.random.rand(5)
random_array
```
