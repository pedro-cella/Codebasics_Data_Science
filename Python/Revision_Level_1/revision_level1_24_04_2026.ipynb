## ⚔️ EXERCISE 7 — Seaborn Scatter Plot

Create a DataFrame:

- hours studied vs exam score

👉 Task:
- Plot using seaborn

👉 Question:
- Is there a positive correlation?

```python
# Definindo uma semente para os números serem sempre os mesmos ao rodar
import numpy as np
np.random.seed(42)

# Criando 20 valores de horas estudadas (entre 1 e 10 horas)
horas_estudadas = np.random.uniform(1, 10, 20)

# Criando o Score com uma relação linear + um pouco de ruído (aleatoriedade)
# A lógica aqui é: Nota = (Horas * 9) + um valor aleatório, limitado a 100
exam_score = (horas_estudadas * 9) + np.random.normal(0, 5, 20)
exam_score = np.clip(exam_score, 0, 100) # Garante que a nota não passe de 100

# Montando o DataFrame
df = pd.DataFrame({
    "Hours_Studied": np.round(horas_estudadas, 1),
    "Exam_Score": np.round(exam_score, 1)
})

df
```

```python
sns.scatterplot(df)
plt.title("Hours Studied vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.show()
```
