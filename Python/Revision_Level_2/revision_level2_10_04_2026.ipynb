## 🧠 Missão 1 — Classificador de Números

Cria um programa que:

- Recebe uma lista de números
- Para cada número:
    - Se for par → printa "Par"
    - Se for ímpar → printa "Ímpar"
    - Se for maior que 100 → printa "Alto" junto

📌 Exemplo esperado:
```python
Entrada: [10, 101, 55]

Saída:
10 → Par
101 → Ímpar e Alto
55 → Ímpar
```


````python
numbers = [10, 101, 55]

print("Saída:")
for number in numbers:
    text = ""
    if number > 100:
        text = "e Alto"
        
    if number % 2 == 0:
        text = "Par " + text
    else:
        text = "Ímpar " + text
    print(f"{number} -> {text}")
````

## ⚙️ Missão 2 — Contador de Palavras

Dada uma frase (string), faça:

- Contar quantas palavras existem
- Criar um dicionário com:
    - palavra → quantidade de vezes que aparece

📌 Exemplo:
```python
"python é legal python é fácil"

Resultado:
{
 "python": 2,
 "é": 2,
 "legal": 1,
 "fácil": 1
}
```

````python
sentence = "três pratos de trigo para três tigres tristes"

# Created an empty dictionary
word_dict = {}

# Converts a String to a List
sentence = sentence.split()

# Iterate over the list to add on the dictionary
for word in sentence:
    # Adds +1 
    if word in word_dict:
        word_dict[word] += 1
    else:
        word_dict[word] = 1

print(word_dict)
````

## 🏹 Missão 3 — Sistema de Notas

Você recebe um dicionário:
```python
alunos = {
    "Ana": 85,
    "João": 92,
    "Maria": 58,
    "Pedro": 76
}
```
Cria um programa que:

- Classifica cada aluno:
    - A (90+)
    - B (80–89)
    - C (70–79)
    - D (60–69)
    - F (<60)
- Retorna um novo dicionário:
```python
{
 "Ana": "B",
 "João": "A",
 "Maria": "F",
 "Pedro": "C"
}
```

````python
students = {
    "Ana": 85,
    "João": 92,
    "Maria": 58,
    "Pedro": 76
}

grades_result = {}

# for student, grade in zip(students.keys(), students.values()):

# Correção
for student, grade in students.items():
    if grade >= 90:
        grades_result[student] = "A"
    elif grade >= 80 and grade <= 89:
        grades_result[student] = "B"
    elif grade >= 70 and grade <= 79:
        grades_result[student] = "C"
    elif grade >= 60 and grade <= 69:
        grades_result[student] = "D"
    elif grade < 60:
        grades_result[student] = "F"

print(grades_result)
````

## 🧪 Missão 4 — Função Inteligente

Cria uma função:
```python3
def analisar_lista(lista):
```
Ela deve retornar:

- Maior valor
- Menor valor
- Média

📌 Saída:
```python
(10, 1, 5.5)
```

````python
def analisar_lista(lista):
    # bigger_number = 0
    # lesser_number = 100

    # Correção
    bigger_number = lista[0]
    lesser_number = lista[0]
    
    total = 0
    for number in lista:
        if number > bigger_number:
            bigger_number = number
        if number < lesser_number:
            lesser_number = number
        total += number
    mean = total/len(lista)
    
    result = (bigger_number, lesser_number, mean)
    
    return result

lista = [2, 3, 5, 1, 9, 18, 9, 34]

print(analisar_lista(lista))
````

## 👑 Missão 5 — Mini Desafio (Combinação)
