## 🧠 Missão 1 — Criando sua primeira classe

Cria uma classe:
```python
class Pessoa:
```

Ela deve ter:

atributos:
- nome
- idade
  
método:
- apresentar() → retorna:
    - "Olá, meu nome é X e tenho Y anos"

````python
class Pessoa:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade

    def apresentar(self):
        return f"Olá, meu nome é {self.nome} e tenho {self.idade}"

pedro = Pessoa("Pedro", 27)
print(pedro.apresentar())
````

## ⚙️ Missão 2 — Sistema com múltiplos objetos

Cria:

uma lista de objetos Pessoa
percorre essa lista e chama apresentar() para cada um

````python
maria = Pessoa("Maria", 34)
ana = Pessoa("Ana", 20)
caio = Pessoa("Caio", 28)
paulo = Pessoa("Paulo", 45)

pessoas = [maria, ana, caio, paulo]
for pessoa in pessoas:
    pessoa.apresentar()
````

## 🏹 Missão 3 — Herança

Cria uma classe:
```python
class Funcionario(Pessoa):
```
Ela deve ter:

- novo atributo:
    - salario
- novo método:
    - mostrar_salario()

````python
class Funcionario(Pessoa):
    def __init__(self, nome, idade,salario):
        super().__init__(nome, idade)
        self.salario = salario

    def mostrar_salario(self):
        return print(f"Salário: {self.salario}")
````

## 🧪 Missão 4 — Sobrescrita de método (override)

Na classe Funcionario:

sobrescreve o método apresentar()
agora deve incluir o salário na mensagem

📌 Exemplo:
```python
Olá, meu nome é Pedro, tenho 25 anos e ganho R$3000
```

````python
class Funcionario(Pessoa):
    def __init__(self, nome, idade, salario):
        super().__init__(nome, idade)
        self.salario = salario

    def mostrar_salario(self):
        return print(f"Salário: {self.salario}")

    def apresentar(self):
        return f"Olá, meu nome é {self.nome}, tenho {self.idade} e ganho R${self.salario}"

teste = Funcionario("Alana", 65, 3500)
print(teste.apresentar())
````

## 👑 Missão 5 — Operator Overloading

Cria uma classe:
```python
class Produto:
```
Com:
- nome
- preco

Agora:
- implementa __add__
- ao somar dois produtos, retorna a soma dos preços

📌 Exemplo:
```python
p1 + p2 → 300
```

````python
class Produto:
    def __init__(self, nome, preco):
        self.nome = nome
        self.preco = preco

    def __add__(self, outro):
        return self.preco + outro.preco

manteiga = Produto("Manteiga", 4.75)
queijo = Produto("Queijo", 5.0)
presunto = Produto("Presunto", 7.0)

print(manteiga + queijo)
````

## ⚔️ Missão 6 — Exception Handling real

Cria uma função:
- recebe um número
- tenta dividir 100 por esse número

Regras:
- se for 0 → tratar erro
- se não for número → tratar erro

📌 Saída:
```python
Erro: divisão por zero
Erro: valor inválido
```

````python
def number_input(number):
    try:
        return 100/number
    except ZeroDivisionError as ze:
        print("Erro: divisão por zero")
    except TypeError as te:
        print("Erro: valor inválido")

#number_input(2)
#number_input(0)
number_input("a")
````

## 🧩 Missão 7 — Sistema completo (mini projeto)

Cria um sistema com:

Classe:
```python
class ContaBancaria:
```

Atributos:
- titular
- saldo

Métodos:
- depositar(valor)
- sacar(valor)
- mostrar_saldo()

Regras:
- Não pode sacar mais do que tem
- Tratar erros com try/except
- Mostrar mensagens claras

💡 Exemplo:
```python
Saldo insuficiente
Depósito realizado
```

````python
class InsufficientFunds(Exception):
    pass

class InvalidValue(Exception):
    pass

class ContaBancaria:
    def __init__(self, titular, saldo):
        self.titular = titular
        self.saldo = saldo

    def depositar(self, valor):
        try:
            if valor <= 0:
              raise InvalidValue  
            self.saldo += valor
            print(f"Depósito realizado")
        except InvalidValue:
            return print("Valor inserido inválido!")

    def sacar(self, valor):
        try:
            if valor > self.saldo:
                raise InsufficientFunds
            elif valor <= 0:
                raise InvalidValue
            self.saldo -= valor
            print(f"Foi realizado o saque de: R${valor}")
            print(f"Saldo atual: R${self.saldo}")
        except InsufficientFunds:
            return print("Operação inválida! Saldo insuficiente")
        except InvalidValue:
            return print("Valor inserido inválido!")
            
    def mostrar_saldo(self):
        return f"Saldo atual: R${self.saldo}"

conta = ContaBancaria("Pedro", 20000)
# conta.sacar(5000)
# conta.depositar(3000)
# conta.mostrar_saldo()
conta.sacar(2000)
conta.mostrar_saldo()
````
