## 🏹 Missão 3 — Sistema com random

Usa o módulo random para:

- Gerar um número aleatório entre 1 e 10
- Criar um jogo:
    - usuário tenta adivinhar
    - printa se acertou ou não

````python
import random

number = random.randint(1, 10)

user_input = int(input("Try to predict the number: "))

if user_input == number:
    print("Congratulations! You won!")
else:
    print("You missed! You lost!")
````
