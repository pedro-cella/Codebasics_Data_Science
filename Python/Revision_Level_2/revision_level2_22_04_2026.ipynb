````python
df = pd.read_csv('../Module_9/Handle_NA_Values/weather_data-Copy1.csv')

# 1. Defina sua lista de valores esquisitos
valores_estranhos = [-99999, -88888, 'no event']

# 2. Filtre o DataFrame original apenas por esses valores
# Usamos .stack() para olhar todas as colunas de uma vez
df_estranhos = df[df.isin(valores_estranhos)]

# 3. Conte as ocorrências
resultado = df_estranhos.stack().value_counts().reset_index()
resultado.columns = ['Valor Esquisito', 'Total de Aparições']

print(resultado)
````

## ⚔️ MISSÃO 2 — Replace estratégico

No dataset de clima (`weather_data`):

👉 Você deve:
- Substituir `-99999` e `-88888` por `NaN`
- Tratar o valor `"no event"`

👉 Perguntas:
- Você substituiria `"no event"` por `NaN` ou por `"Sunny"`?
- Em que cenário cada decisão faz sentido?

👉 **Entrega:**
- Decisão + justificativa

````python
df2 = df.copy()
df2
````

````python
df2 = df2.replace({
    -99999.0: np.nan,
    -88888.0: np.nan,
    r'^\s*no event\s*$': np.nan
}, regex=True)
df2
````

If we treat this as a real world dataset, I would change "no event" for NaN, because I don't want to interefere with the data without knowing its consequences as changing 'no event' for "Sunny" could cause.
