# ⚔️ MISSÃO 1 — Basic Concat

👉 Combine two small datasets:

Create:
- df_online_orders (first 2 rows from orders)
- df_store_orders (last 2 rows from orders)

👉 Task:
- Concatenate them vertically

👉 Questions:
- What happens to the index?
- How to fix it?

```python
df = pd.read_csv('orders.csv')
df
```

```python
df_online_orders = df.head(2)
df_online_orders
```

```python
df_store_orders = df.tail(2)
df_store_orders
```

```python
df_orders_concat = pd.concat([df_online_orders, df_store_orders], ignore_index=True)
df_orders_concat
```

The indexes won't follow an order, the way to fix it is to use the argument `ignore_index` that way the index will follow the correct numerical order

# ⚔️ MISSÃO 2 — Horizontal Concat

Create:
- df_customer_names → only name column
- df_customer_cities → only city column

👉 Task:
- Concatenate horizontally (axis=1)

👉 Question:
- What happens if indexes don’t match?

```python
df2 = pd.read_csv('customers.csv')
df2
```

```python
df_customer_names = df2[['name']]
df_customer_names
```

```python
df_customer_cities = df2[['city']]
new_row = pd.DataFrame({'city': 'Washigton D.C'}, index=[4])
df_customer_cities = pd.concat([df_customer_cities, new_row], ignore_index=True)
df_customer_cities
```

```python
df_customers_name_city = pd.concat([df_customer_names, df_customer_cities], axis=1)
df_customers_name_city
```

When using `.concat()` if an index does not match, it creates a new row, with one of the values, with the dataframe that does not have a matched index, as NaN.

# ⚔️ MISSÃO 3 — Inner Join

👉 Merge:
- orders + customers

👉 Task:
- Use INNER JOIN on customer_id

👉 Questions:
- Which rows disappeared?
- Why?

```python
df_orders = pd.read_csv('orders.csv')
df_orders
```

```python
df_customers = pd.read_csv('customers.csv')
df_customers
```

```python
df_customers_orders = pd.merge(df_orders, df_customers, on='customer_id', how='inner')
df_customers_orders
```

from `orders.csv` the last row, index 3. and from `customers.csv` the rows 3 and 4, indexes 2 and 3,  disappeared, the motive is in the `inner` argument, when using this arugment we use an INNER JOIN just like in SQL, so only the `customer_id` that appeared in both data would be included in the merge.
