## ⚔️ EXERCISE 1 — List Comprehension with Condition

Create a list:

numbers = [1, 2, 3, 4, 5, 6, 7, 8]

👉 Task:
- Create a new list with only EVEN numbers
- Square those numbers

````python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

squared_even_numbers = [number*number for number in numbers if number % 2 == 0]

print(squared_even_numbers)
````

## ⚔️ EXERCISE 2 — Conditional Transformation

Given:

prices = [120, 80, 150, 60, 200]

👉 Task:
- Apply a 10% discount ONLY to prices greater than 100
- Use list comprehension

````python
prices = [120, 80, 150, 60, 200]

discount = [price - price*0.1 if price > 100 else price for price in prices]

discount
````

## ⚔️ EXERCISE 3 — Dictionary Comprehension

Given:

names = ["Alice", "Bob", "Charlie"]
scores = [85, 90, 78]

- Create a dictionary mapping name → score

````python
names = ["Alice", "Bob", "Charlie"] 
scores = [85, 90, 78]

name_score_dict = {name:score for name, score in zip(names, scores)}
name_score_dict
````

## ⚔️ EXERCISE 4 — Set Comprehension

Given:

```python
transactions = [
    {'customer': 'A', 'amount': 100},
    {'customer': 'B', 'amount': 200},
    {'customer': 'A', 'amount': 150}
]
```

- Extract UNIQUE customers using set comprehension

````python
transactions = [
    {'customer': 'A', 'amount': 100},
    {'customer': 'B', 'amount': 200},
    {'customer': 'A', 'amount': 150}
]

unique_customers = {customer['customer'] for customer in transactions}
unique_customers
````

## ⚔️ EXERCISE 5 — Set Operations

Given:
```python
a = {"Alice", "Bob", "Charlie"}
b = {"Bob", "David", "Emma"}
```
👉 Tasks:
- Find intersection
- Find union
- Find difference (a - b)

👉 Question:
- What does each result represent?

````python
a = {"Alice", "Bob", "Charlie"}
b = {"Bob", "David", "Emma"}

intersection = a.intersection(b)
union = a.union(b)
diff = a.difference(b)

print(intersection)
print(union)
print(diff)

# 1. It saves the value that appear in both sets "a" and "b"
# 2. It saves all values from "a" and "b" 
# 3. Show only the values that appear in "a" but not in "b"
````
