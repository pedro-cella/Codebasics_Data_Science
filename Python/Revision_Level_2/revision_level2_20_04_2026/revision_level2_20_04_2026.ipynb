```python
df.nlargest(5, 'profit')
```

About the question, no revenue and profit are not always the same thing. When talking about movies or the success of a movie, you must think about 3 variables: budget, revenue and profit. Budget is how much a movie cost to be made, Revenue usually means how much a movie could pay itself, that's why in some cases there are minimal revenue costs that must be targeted, and profit is about how much of revenue we have made over the amount we needed to make the movie. Basically budget means the cost of the movie, revenue means the money that pays the movie, and profit is the amount gain beyond the cost of the movie.

## 6. Análise por faixa de ano

Agrupe por:

`year_classify`

E calcule:

- média de budget
- média de revenue
- média de profit
- quantidade de filmes

```python
df.groupby('year_classify').agg({
    'budget': 'mean',
    'revenue': 'mean',
    'profit': 'mean',
    'title': 'count'
})
```

## 8. Eficiência financeira

Crie uma nova coluna chamada algo como:

`roi`
ou
`return_ratio`

Onde tu vai comparar retorno em relação ao orçamento.

A ideia é descobrir:

- quais filmes foram mais eficientes financeiramente

Depois mostre os 5 melhores.
