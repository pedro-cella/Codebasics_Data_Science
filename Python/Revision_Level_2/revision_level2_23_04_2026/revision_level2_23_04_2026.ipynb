## ⚔️ MISSÃO 1 — Receita por região

👉 Crie uma coluna:
revenue = price * quantity

Depois:

👉 Agrupe por `region` e calcule:
- receita total
- quantidade total vendida

👉 Pergunta:
- qual região gera mais receita?
- isso está ligado à quantidade ou ao preço?


```python
df = pd.read_csv('ecommerce_orders.csv')
df
```

```python
df.shape
```

```python
df.size
```

```python
df['revenue'] = df['price']*df['quantity']
df
```

```python
df.groupby('region').agg({
    'revenue': 'sum',
    'quantity': 'sum',
    'price': 'sum'
})
```

The region that contributed more in revenue was the North region, when comparing the price as well, we can see that both quantity an price in the North region havean impact on the revenue, but we could not saythat quantity interferes so much in revenue, since the South region is the second in quantity but the last in revnue, so based only on the columns: revenue, quantity and price, we can say that quantity does not interefere as much as price.

## ⚔️ MISSÃO 3 — Cliente mais valioso
