## ⚔️ EXERCISE 5 — Custom Iterator

Create a class:

👉 Task:
- Create a class that returns numbers from 1 to 3
- Implement __iter__ and __next__

👉 Goal:
- Use a for loop to iterate over it

```python
class ReturnNumbers:
    def __init__(self):
        self.limit = 3
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current <= self.limit:
            val = self.current
            self.current += 1
            return val
        else:
            raise StopIteration

numbers = ReturnNumbers()
for _ in range(10):
    print(next(numbers))
```

## ⚔️ EXERCISE 6 — Generator Basics

👉 Task:
- Create a generator function that yields numbers from 1 to 5

👉 Use:
- next() manually
- a for loop

👉 Question:
- What is the difference between generator and list?

```python
def numbers_from_1_to_5():
    current = 1
    while current <= 5:
        yield current
        current += 1
    else:
        raise StopIteration

g = numbers_from_1_to_5()
# print(next(g))
# print(next(g))
# print(next(g))
# print(next(g))
# print(next(g))
# print(next(g))

for _ in range(6):
    print(next(g))
```

When comparing a List and a Generator we can focus on how the iteration occurs on both, while on a List, iterating over the elements using a for loop, will print all elements in one go, of course it can change depending on the if condition, but it will be executed once, while a generator, that can only be a generator if yield is used, can separate separated this elements, that's because when using yield, it pauses the execution send the current state, return a value to its caller, and then can be resumed later. just like a game checkpoint.

## ⚔️ EXERCISE 7 — Generator with Condition

👉 Task:
- Create a generator that yields only EVEN numbers from 1 to 10

👉 Goal:
- Understand filtering with yield

```python
def even_numbers():
    current = 1
    while current <= 10:
        if current % 2 != 0:
            yield current
            current += 1
        else:
            current += 1

even = even_numbers()
for _ in range(10):
    print(next(even))
```

## ⚔️ EXERCISE 8 — Real Scenario Generator

👉 Task:
- Create a generator that reads a list of temperatures
- Yield only temperatures above 25

Example:
temps = [22, 28, 30, 19, 35]

```python
temps = [22, 28, 30, 19, 35]

def check_temperature(temps):
    for temp in temps:
        if temp > 25:
            yield temp

check = check_temperature(temps)
for t in check:
    print(t)
```

## ⚔️ EXERCISE 9 — Decorator Basics

👉 Task:
- Create a decorator that prints:
  "Function started" and "Function ended"

👉 Apply it to a function that prints a message

```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Function started")

        result = func(*args, **kwargs)

        print("Function ended")
        return result
    return wrapper
```

```python
@my_decorator
def print_message(name):
    return f"Hello World! {name}"

print_message("Pedro")
```

## ⚔️ EXERCISE 11 — Cache Decorator (IMPORTANT)

👉 Task:
- Create a decorator that caches results of a function

Example:
get_square(4) → compute  
get_square(4) → return cached  

👉 Goal:
- Understand real-world use of decorators


```python
def cache_decorator(func):
    cache = {}
    def wrapper(*args, **kwargs):
        key = args

        if key in cache:
            print(f"Buscando do cache para: {key}")
            return cache[key]
        else:
            print(f"Calculando pela primeira vez para: {key}")
            result = func(*args, **kwargs)
            cache[key] = result
            
        return result
    return wrapper
```

```python
@cache_decorator
def get_squared(number):
    return number*number

print(get_squared(4))
print(get_squared(4))  # Vai buscar no cache
print(get_squared(5))  # Vai calcular
```
