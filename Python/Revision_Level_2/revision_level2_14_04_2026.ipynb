## 🧠 Missão 18 — Estatística prática

Dado:
```python
[1200, 1500, 800, 2000, 950]
```
- média
- maior valor
- quantos estão abaixo da média

````python
array_mission_18 = np.array([1200, 1500, 800, 2000, 950])
# Media
print(f"Média: {np.mean(array_mission_18)}")
# Maior Valor
print(f"Maior Valor: {np.max(array_mission_18)}")
# Quantos estão abaixo da média
print(f"Abaixo da média: {(array_mission_18 < np.mean(array_mission_18)).sum()}")
````

## 👑 Missão 19 — Linha com maior soma

Dado:
```python
[
 [1, 2, 3],
 [10, 20, 30],
 [5, 5, 5]
]
```
qual linha tem a maior soma?

````python
matrix_mission_19 = np.array([
     [1, 2, 3],
     [10, 20, 30],
     [5, 5, 5]
])
# Minha tentativa
#print(np.max(np.sum(matrix_mission_19, axis=1)))
# Resposta correta
print(np.argmax(np.sum(matrix_mission_19, axis=1)))
````
