## ⚔️ EXERCISE 2 — Handling Status Code

👉 Task:

- If status_code == 200 → print "Success"
- Else → print "Error"

👉 Goal:

```python
url = "https://jsonplaceholder.typicode.com/posts/1"

response = requests.get(url)

if response.status_code == 200:
    print("Success")
else:
    print("Error")
```

## ⚔️ EXERCISE 4 — POST Request

👉 Task:
- Send a POST request to:
  https://jsonplaceholder.typicode.com/posts

👉 Data:
{
  "title": "test",
  "body": "learning API",
  "userId": 1
}

👉 Tasks:
- Print status code
- Print response JSON

👉 Question:
- What status code did you get? Why?

```python
url = "https://jsonplaceholder.typicode.com/posts"
data = { "title": "test", "body": "learning API", "userId": 1 }

response = requests.post(url, json=data)

print(response.status_code)

data = response.json()
print(data)
```

I got a status code of 201, which indicates that the request succeeded and created a new resource.

## ⚔️ REQUIREMENTS

Inside the function:

1. Make a GET request to:
   https://jsonplaceholder.typicode.com/posts

2. Check if status_code == 200

3. Convert response to JSON

4. Print:
   - title
   - body

```python
def get_posts(number_of_posts):
    url = "https://jsonplaceholder.typicode.com/posts"
    response = requests.get(url)
    posts_list = []
    try:
        posts = response.json()
        for post in posts[:number_of_posts]:
            posts_list.append(post)
        return posts_list
    except Exception as err:
        return f"An error occurred {err}"

posts = get_posts(3)
for post in posts:
    print(f"ID: {post['id']}\nTitle: {post['title']}\nBody: {post['body']}\n")
```
