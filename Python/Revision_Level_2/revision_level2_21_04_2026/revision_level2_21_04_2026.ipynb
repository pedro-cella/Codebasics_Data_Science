In this case I used multiple methods for fill na values on the Dataframe, 

I interpolate on values from columns: `units_sold` and `price`, that way the result would not create a possible outlier or when doing the mean or median, the values would still follow a pattern.

For the `temperature` and `rain_mm` I used the .fillna() and calculated the mean(), that way both would be values that are similarly close to other values.

For `promo` I used .ffill() but if we had another column that could explain it when do we consider a promo or note we could have filled with real values

For `holiday` i used the bfill(), but similarly from the case before, we could check if that row date is a holiday or not so that way we could have replaced the null values for real world values, the problem is that I do not know what country those days are based on

## ⚔️ MISSÃO 4 — Aplicação controlada

Agora sim:

👉 Aplique suas decisões  

**Regras:**
- Não alterar o dataset original  
- Criar uma versão limpa  

👉 **Entrega:**
- DataFrame limpo  

```python
# Fill na by interpolation
df_clean = df.copy()
df_clean.units_sold = df_clean.units_sold.interpolate()
df_clean.price = df_clean.price.interpolate()

# Fill na by ffill
df_clean.promo = df_clean.promo.ffill()

# Fill na by bfill
df_clean.holiday = df_clean.holiday.bfill()

# Fill na by using fillna with mean()
df_clean = df_clean.fillna({
    'temperature': df_clean.temperature.mean(),
    'rain_mm': df_clean.rain_mm.mean()
})

df_clean = df_clean.round({
    'units_sold': 1, 
    'price': 1, 
    'temperature': 1, 
    'rain_mm': 1
})

df_clean
```
