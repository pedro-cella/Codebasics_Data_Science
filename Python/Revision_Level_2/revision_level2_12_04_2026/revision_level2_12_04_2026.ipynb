## 🧠 Missão 1 — Leitor de Arquivo

Cria um arquivo:
```python
dados.txt
```

Com conteúdo exemplo:
```python
10
20
30
40
```

Agora faz um programa que:

- Lê o arquivo
- Converte os valores para número
- Soma tudo
- Mostra o resultado

📌 Esperado:
```python
Total: 100
```

````python
# Creating a file with "w"
with open("dados.txt", "w") as file:
    file.writelines([
        "10\n",
        "20\n",
        "30\n",
        "40\n"
    ])
````

````python
# Create an empty list
numbers = []

# Read file
with open("dados.txt", "r") as file:
    for line in file:
        numbers.append(int(line.strip()))

print(f"Total: {sum(numbers)}")
````

## 🧪 Missão 4 — Filtro de Dados

Cria um arquivo com números misturados:
```python
5
12
7
30
2
```
Agora:

- Lê o arquivo
- Cria um novo arquivo maiores_que_10.txt
- Salva apenas os números maiores que 10

📌 Resultado esperado:
```python
12
30
```

````python
# Creating file
with open("random_numbers.txt", "w") as file:
    file.writelines([
        "5\n",
        "12\n",
        "7\n",
        "30\n",
        "2\n"
    ])
````

````python
# Read first file
with open("random_numbers.txt", "r") as file:
    # Iterate throw line to line
    for line in file:
        # Transforms a line into a integer and validate the conditional
        if int(line.strip()) > 10:
            # Creates a new file and add a line if the condition is true
            with open("maiores_que_10.txt", "w") as file2:
                file2.write(line)
````

## 👑 Missão 5 — Sistema de Log (mundo real)

Cria um sistema que:

- Toda vez que rodar, salva no arquivo log.txt:
    - Data atual
    - Mensagem: "Programa executado"

📌 Exemplo:
```python
2026-04-12 10:30 - Programa executado
```

````python
from datetime import datetime

now = datetime.now()
current_date = now.date()
current_hour = now.hour
current_minute = now.minute

with open("log.txt", "w") as file:
    file.write(f"{current_date} {current_hour}:{current_minute} - Programa executado")
````

## 🔥 Missão 7 — Mini Sistema (Combinação)

Cria um sistema que:

- Usuário digita um nome
- Salva no arquivo
- Depois lê o arquivo
- Mostra todos os nomes já salvos

````python
# User insert a name
user_input = input("Write a name: ")

# Save the name on a txt file
with open("names_file.txt", "a") as file:
    file.write(f"{user_input}\n")

# Read the file
with open("names_file.txt", "r") as file:
    # Show all names saved
    print("Names saved: ")
    for line in file:
        print(line.strip())
````
