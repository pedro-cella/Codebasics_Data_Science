# Modules and Pip
- A module is a python code returned by somebody else that can be reused
- To install some module just use `pip install _`
- To uninstall some module just use `pip uninstall _`
- To look at the version of the module use `pip freeze | grep numpy`
- To import an specific function use from `module` import `function`
- And to import the whole module, just use `import module`

## Takeaways
- Modules in Python allow you to **organize** and **reuse code** across multiple programs
- You can import built-in modules like **math** and **external modules** which are available on **https://pypi.org/**
- Use **import module_name** to bring a module into your current script
- **The pip tool** is used to install and manage external Python packages
- **USe pip install package_name** to add new packages to your Python environment
