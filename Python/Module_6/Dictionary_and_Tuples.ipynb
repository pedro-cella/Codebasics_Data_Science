```python
def find_pe_and_pb(price, eps, book_value):
    pe = price/eps
    pb = price/book_value
    return pe, pb
```

## Takeaways
- **Tuples** are immutable data structures in Python, which means **once a tuple is created,** it's elements cannot be ***changed, added** or **removed.**
- **Dictionaries in Python** are mutable data structures that store key-value pairs, allowing fast data retrieval by key.
- Dictionary is a Python-specific implementation of a **HashMap** data structure.
- Keys in dictionaries must be unique and immutable types, such as **strings, numbers,** or **tuples,** ensuring that each key maps distinctly to its value.
- Both **tuples** and **dictionaries support various methods** and operations such as accessing **elements, iterating over them, etc** for efficient data manipulation and retrieval.
