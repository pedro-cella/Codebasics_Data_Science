```python
# functions_expenses
def find_total(expenses):
    '''
    :param expenses: input list containing numbers
    :return: total sum of all numbers in the input list
    '''
    total = 0
    for expense in expenses:
        total += expense

    return total
```

```python
expenses_sergey = [30, 45, 70, 90]
expenses_sundar = [40, 23, 10, 85]

total_expense_sergey = find_total(expenses_sergey)
# for expense in expenses_sergey:
#     total_expense_sergey += expense
print("Total surgey's expense:", total_expense_sergey)

total_expense_sundar = find_total(expenses_sundar)
# for expense in expenses_sundar:
#     total_expense_sundar += expense
print("Total surgey's expense:", total_expense_sundar)
```

```python
# cylinder_volume

def find_cylinder_volume(radius, height=7):
    print("radius:", radius)
    print("height:", height)
    volume = 3.14*(radius**2)*height
    return volume

r = 10

v = find_cylinder_volume(height=67, radius=r)
print(v)
```

```python
# function_arguments
def sum_all(*args):
    total = 0
    for num in args:
        total += num
    return total

total = sum_all(1, 2, 3, 4, 5, 6)
print(total)
```

# Takeaways
- Functions in Python are defined using the **def keyword,** followed by a **function name and paentheses,** which may include parameters
- **Parameters** allow functions to receive input making them flexible and reusable for different data inputs
- The **return statement** is used to send a function's result back to the caller, exiting the function and optionally passing back a value
- **Default parameters** in function definitions provide default values for arguments, simplifying function calls and enhancing flexibility
- Python supports **both positional and keyword arguments,** giving more control over how functions are called and how arguments are passed to them
- **lambda** is a quick way to define a function in a single line
