# File Handling
- .close() dealocates resources that were allocated when the file was opened.
- using `with` crates a context which when the context finishes it will automatically calls the .close() to clean the allocated resources.
- We can use `_` to ignore a value, an example on this lesson while saving data on the variables player, match_id, score = data, the value of match_id was useless, so was substituted for player, _, score

```python
f = open("funny.txt", "r")

for line in f:
    print(line)

f.close()
```

```python
with open("funny.txt", "r") as f:
    for line in f:
        print(line)
```

```python
with open("funny.txt", "r") as f:
    lines = f.readlines()
    for line in lines:
        print(line)
```

```python
with open("love.txt", "w") as f:
    f.write("I love Python\n")
    f.write("I love Meditation")
```

```python
player_scores = {}

with open("scores.csv", "r") as f:
    for line in f:
        player, _, score = line.split(",")
        score = int(score.strip())
        if player in player_scores:
            player_scores[player].append(score)
        else:
            player_scores[player] = [score]

player_scores
```

```python
int('9\n'.strip())
```

```python
for player, scores_list in player_scores.items():
    min_scores = min(scores_list)
    max_scores = max(scores_list)
    avg_scores = sum(scores_list)/len(scores_list)
    print(f"{player} ==> Min: {min_scores}, Max: {max_scores}, Avg: {avg_scores}")
```

## Takeaways
- The key function for working with files in Python is the **open()** function. The **open()** function takes two parameters; **filename,** and **mode.**
- Here are gour different methods(modes) for opening a file:
    - **"r"** - Read - Default value. Opens a file for reading, error if the file does not exist
    - **"a"** - Append - Opens a file for appending, creates the file if it does not exist
    - **"w"** - Write - Opens a file for writing, creates the file if it does not exist
    - **"x"** - Create - Creates the specified file, returns an error if the file exist
- It is good practive to use **with** keyword when dealing with file objects. The advantage is that the file is properly closed after its suite finishes, even if an exception is raised at some point.
