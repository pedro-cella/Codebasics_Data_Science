## Takeaways:
- **Variables** are containers that store data in a Python program;
- Variablescan store **different types of data types** such as string, integer, float, boolean;
- Rules of naming variables correctly:
  - You **can not use reserve words** as variables names (e.g. dev, True, etc.)
  - Variables names **must start with a letter or an underscore _**;
  - **Underscore _ is a valid character** and can be present at any place in a variable name;
