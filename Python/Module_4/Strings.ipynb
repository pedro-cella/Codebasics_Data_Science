## Takeaways
- In Python, strings are immutable, meaning they cannot be changed once created.
- Access specific characters or substrings in Python using indexing (e.g., **name[0]**) and slicing (e.g., **name[1:5]**).
- Use string formatting techniques like f-strings(e.g, **f'{variable}'**) for easier and more readable string compsition in Python.
- Python provides built-in methods for strings, such as **.upper(), .lower(), .split(),** and **.strip()**
